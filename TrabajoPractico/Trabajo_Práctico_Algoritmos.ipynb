#Modelo
---
- ¿Como represento el espacio de soluciones?
  
El espacio de soluciones está dado por todas las posibles asignaciones de las 30 tomas a días de grabación, respetando las restricciones de que:

- No se pueden grabar más de 6 tomas por día.
- Un actor solo debe presentarse en el estudio si participa en al menos una de las tomas del día.

Definimos las variables de decisión:

- $x_{t,d}$= 1 si la toma $t$ se asigna al día $d$, 0 en caso contrario.
- $y_{a,d}$ = 1 si el actor $a$ trabaja el día $d$, 0 en caso contrario.

El conjunto de soluciones posibles está dado por todas las combinaciones en las que se asignan las tomas a los días y se determina qué actores asisten cada día.

Para representar los datos y resultados, se emplearán estructuras de datos en Python, como listas y DataFrames de Pandas.

---
- ¿Cual es la función objetivo?
  
El objetivo es minimizar el costo total de los actores, es decir, minimizar la cantidad de días en los que cada actor debe ir a grabar.

$$
\text{Costo total} = min\sum_{d=1}^{D} \sum_{a=1}^{10} \left( y_{a,d} \right)
$$

Donde:
- $D$ es el número total de días de grabación (desconocido a priori, pero no mayor a 30/6 = 5 días).
- Cada $y_{a,d}$ representa si el actor $a$ tiene que presentarse en el día $d$.

Al minimizar esta suma, buscamos reducir la cantidad de días que se le paga a cada actor.


---
- ¿Como implemento las restricciones?
  
  1. Cada toma debe ser asignada exactamente a un día:
   $$
   \sum_{d=1}^{D} x_{t,d} = 1, \quad \forall t \in \{1, \dots, 30\}
   $$

    Es decir, cada toma debe ser asignada a un único día.
  2. No se pueden grabar más de 6 tomas por día:
   $$
   \sum_{t=1}^{30} x_{t,d} \leq 6, \quad \forall d \in \{1, \dots, D\}
   $$
  3. Si un actor participa en una toma asignada a un día, debe asistir ese día:
   $$
   y_{a,d} \geq x_{t,d}, \quad \forall t \in \{1, \dots, 30\}, \forall a \in A_t, \forall d \in \{1, \dots, D\}
   $$
    Donde $A_t$ es el conjunto de actores que participan en la toma $t$. Esto asegura que si un actor está en una toma ese día, se active su asistencia.
  4. Las variables deben ser binarias:
   $$
   x_{t,d} \in \{0,1\}, \quad y_{a,d} \in \{0,1\}
   $$

    Esto garantiza que cada toma solo se asigne a un día y que los actores solo sean convocados cuando sea necesario.

#Análisis
- ¿Que complejidad tiene el problema?. Orden de complejidad y Contabilizar el espacio de soluciones

El problema de **planificación óptima de sesiones de doblaje** es un caso del **problema de optimización combinatoria**, similar al **bin packing problem** o **scheduling problem** con restricciones adicionales. Se puede modelar como un **problema de programación entera (ILP - Integer Linear Programming)**, que en general es **NP-hard**.  

---

## Orden de Complejidad
La complejidad del problema depende de la cantidad de **tomas** y **actores**, y la forma en que se agrupan las tomas en días minimizando costos.

- **Variables de decisión**:  
  - $x_{t,d}$ (asignación de tomas a días) → $ 30D $ variables binarias  
  - $ y_{a,d} $ (asistencia de actores a días) → $ 10D $ variables binarias  
  Donde $ D $ es el número máximo de días (a lo sumo 5 días, ya que $ \lceil 30/6 \rceil = 5 $).  

- **Restricciones**:  
  - 30 restricciones para que cada toma se asigne a un único día.
  - $ D $ restricciones para no superar 6 tomas por día.
  - $ 30 \times 10 \times D $ restricciones para garantizar que un actor esté presente si participa en la toma del día.

La **complejidad del espacio de soluciones** es **exponencial**, ya que involucra decisiones binarias ($ x_{t,d}, y_{a,d} $), lo que sugiere que hay:
$2^{O(30D)}$ combinaciones posibles.

En términos generales, los problemas de **asignación óptima de recursos** como este tienen **complejidad exponencial** o **pseudo-polinomial**, lo que hace que resolverlos exactamente sea costoso para tamaños grandes.  

---

## Tamaño del Espacio de Soluciones
Para estimar el **espacio de soluciones**, consideramos:

1. Cada toma debe asignarse a uno de $ D $ días. Hay $ D^{30} $ formas de hacerlo.  
2. Para cada día, cada actor puede estar presente o no. Como hay 10 actores y máximo $ D $ días, hay $ 2^{10D} $ combinaciones posibles.

El **espacio total de soluciones** (aproximado) es:

$$
S = D^{30} \times 2^{10D}
$$

Para $ D = 5 $, esto da aproximadamente:

$$
5^{30} \times 2^{50} \approx 9.3 \times 10^{22}
$$

Este número es enorme, lo que confirma que **una búsqueda exhaustiva es inviable** y se requieren técnicas de **optimización heurísticas o métodos exactos eficientes** como **Branch & Bound, programación entera mixta (MILP) o algoritmos metaheurísticos (Genéticos, Tabu Search, etc.)**.

---

## Conclusión
- El problema es **NP-hard** debido a la asignación combinatoria con restricciones.  
- La **complejidad exacta** es exponencial en el peor caso \( 2^{O(30D)} \).  
- **Espacio de soluciones**: inmanejable para búsqueda exhaustiva.  
- Es recomendable **programación entera mixta (MILP)** o **heurísticas** para resolverlo en tiempos razonables.

#Diseño
- ¿Que técnica utilizo? ¿Por qué?
  
  La programación lineal entera (PLIE) es la técnica más adecuada para este problema. Esto se debe a que:

  - Permite modelar eficientemente el problema como una asignación de tareas con restricciones.
  - Ofrece una solución exacta, garantizando encontrar la mejor asignación posible.
  - Los solvers de programación lineal (como PuLP en Python) son capaces de resolver problemas de tamaño moderado como el de este caso de manera eficiente.
---

```
#Respuesta

# Número de tomas y actores
num_tomas = df.shape[0]
num_actores = df.shape[1]

# Crear el problema de programación lineal entera
prob = pulp.LpProblem("Planificación_Sesiones_Doblaje", pulp.LpMinimize)

# Variables de decisión: x_t_d = 1 si la toma t se graba en el día d, 0 si no.
x = pulp.LpVariable.dicts("x", (range(num_tomas), range(num_tomas)), cat='Binary')

# Variable para indicar si el actor a trabaja el día d
y = pulp.LpVariable.dicts("y", (range(num_actores), range(num_tomas)), cat='Binary')

# Función objetivo: Minimizar el número total de días en los que los actores trabajan
prob += pulp.lpSum([y[a][d] for a in range(num_actores) for d in range(num_tomas)])

# Restricción 1: Cada toma debe ser asignada a un único día
for t in range(num_tomas):
    prob += pulp.lpSum([x[t][d] for d in range(num_tomas)]) == 1

# Restricción 2: No más de 6 tomas por día
for d in range(num_tomas):
    prob += pulp.lpSum([x[t][d] for t in range(num_tomas)]) <= 6

# Restricción 3: Cada actor solo puede ser asignado a las tomas en las que está involucrado
for a in range(num_actores):
    for d in range(num_tomas):
        prob += pulp.lpSum([x[t][d] for t in range(num_tomas) if df.iloc[t, a] == 1]) >= y[a][d]

# Resolver el problema
prob.solve()

# Inicializar la lista para almacenar los días de grabación
dias_grabacion = [-1] * num_tomas  # Lista para almacenar el día en que se graba cada toma

# Imprimir la asignación y guardar los resultados en la lista
if pulp.LpStatus[prob.status] == 'Optimal':
    for d in range(num_tomas):
        #print(f"Día {d+1}:")
        for t in range(num_tomas):
            if pulp.value(x[t][d]) == 1:
                dias_grabacion[t] = d + 1  # Asignar el día en que se graba la toma t
                #print(f"  Toma {t+1}")

# Crear el DataFrame con la asignación de días de grabación
df_dias_grabacion = pd.DataFrame(dias_grabacion, columns=['Día Grabación'], index=[f'Toma {i+1}' for i in range(num_tomas)])

# Filtrar el DataFrame para mostrar solo las tomas que tienen asignado un día de grabación
df_dias_grabacion = df_dias_grabacion[df_dias_grabacion['Día Grabación'] != -1]

# Mostrar el DataFrame con la asignación de días de grabación
print("\nAsignación de días de grabación:")
print(df_dias_grabacion)

# Eliminar las filas del DataFrame donde no haya grabación
df_dias_grabacion_final = df_dias_grabacion.dropna()
```

```
# Contar el número de días únicos en los que se graba
num_dias_grabacion = df_dias_grabacion_final['Día Grabación'].nunique()

# Mostrar el número de días de grabación
print(f"\nNúmero de días en los que se graba: {num_dias_grabacion}")
```
